| **8-bit Floating Point**  | `torch.float8`    | Ultra-low-precision floating point. Used for experimental applications and extreme memory-constrained environments (less common).                                               |
| **8-bit Integer**         | `torch.int8`      | 8-bit signed integer. Used for quantized models to save memory and computation in inference.                                                                                   |
| **16-bit Integer**        | `torch.int16`     | 16-bit signed integer. Useful for special numerical tasks requiring intermediate precision.                                                                                    |
| **32-bit Integer**        | `torch.int32`     | Standard signed integer type. Commonly used for indexing and general-purpose numerical tasks.                                                                                  |
| **64-bit Integer**        | `torch.int64`     | Long integer type. Often used for large indexing arrays or for tasks involving large numbers.                                                                                  |
| **8-bit Unsigned Integer**| `torch.uint8`     | 8-bit unsigned integer. Commonly used for image data (e.g., pixel values between 0 and 255).                                                                                    |
| **Boolean**               | `torch.bool`      | Boolean type, stores `True` or `False` values. Often used for masks in logical operations.                                                                                      |
| **Complex 64**            | `torch.complex64` | Complex number type with 32-bit real and 32-bit imaginary parts. Used for scientific and signal processing tasks.                                                               |
| **Complex 128**           | `torch.complex128`| Complex number type with 64-bit real and 64-bit imaginary parts. Offers higher precision but uses more memory.                                                                 |
| **Quantized Integer**     | `torch.qint8`     | Quantized signed 8-bit integer. Used in quantized models for efficient inference.                                                                                              |
| **Quantized Unsigned Integer** | `torch.quint8` | Quantized unsigned 8-bit integer. Often used for quantized tensors in image-related tasks.                                                                                     |


# ****Mathematical Operations****

## ****1. Scalar Operations****

```python
z = torch.rand(5, 5)
z
```

```python
# ADDITION
print('Addition --> \n', z + 3)

# SUBSTRACTION
print('\nSubstraction --> \n', z - 1)

# MULTIPLICATION
print('\nMultiplication --> \n', z * 4)

# DIVISION
print('\nDivision -- > \n', z / 3)

# INT DIVISION
print('\nInt Division --> \n', (z * 100) // 3)

# MOD
print('\nMod --> \n', ((z * 100) // 3) % 2)

# POWER
print('\nPower --> \n', z ** 2)
```

## ****2. Element Wise operations****

```python
a = torch.rand(3, 4)
b = torch.rand(3, 4)
c = torch.tensor([-3, 5, -2, 8, -9, 1, -7])


display(a)
display(b)
display(c)
```

```python
# ADDITION
print('Addition --> \n', a + b)

# SUBSTRACTION
print('\nSubstraction --> \n', a - b)

# MULTIPLICATION
print('\nMultiplication --> \n', a * b)

# DIVISION
print('\nDivision -- > \n', a / b)

# MOD
print('\nMod --> \n', (a % b))

# POWER
print('\nPower --> \n', a ** b)

# ABSOLUTE
print('\nAbsolute --> \n', torch.abs(c))

# NEGATIVE
print('\nNegative --> \n', torch.neg(c))

# ROUND
print('\nRound --> \n', torch.round(a))

# CEIL
print('\nCeil --> \n', torch.ceil(b))

# FLOOR
print('\nFloor --> \n', torch.floor(b))

# CLAMP
print('\nClamp --> \n', torch.clamp(a, min = 0.1, max = 0.8))
```

## ****3. Reduction Operations****

```python
d = torch.randint(size = (3, 4), low = 0, high = 10, dtype = torch.float64)
d
```

```python
# SUM
print('\nSum --> \n', torch.sum(d))

# SUM ALONG COLUMNS
print('\nSum Along Columns --> \n', torch.sum(d, dim = 0))

# SUM ALONG ROWS
print('\nSum Along Rows --> \n', torch.sum(d, dim = 1))
```

```python
# MEAN
print('\nMean --> \n', torch.mean(d))

# MEAN ALONG COLUMNS
print('\nMean Along Columns --> \n', torch.mean(d, dim = 0))
```

```python
# MEDIAN
print('\nMedian --> \n', torch.median(d))

# MAX & MIN
print('\nMax --> \n', torch.max(d))
print('\nMin --> \n', torch.min(d))

# PRODUCT
print('\nProduct --> \n', torch.prod(d))

# STANDARD DEVIATION
print('\nStandard Deviation --> \n', torch.std(d))

# VARIANCE
print('\nVariance --> \n', torch.var(d))

# ARGMAX
print('\nArgmax --> \n', torch.argmax(d))

# ARGMIN
print('\nArgmin --> \n', torch.argmin(d))
```

## ****5. Comparison Operations****

```python
h = torch.randint(size = (5, 6), low = 0, high = 30)
i = torch.randint(size = (5, 6), low = 10, high = 50)

print("h --> \n", h)
print("\ni --> \n",i)
```

```python
# GREATER THAN
print('\nGreater than --> \n', h > i)

# LESS THAN
print('\nLess than --> \n', h < i)

# EQUAL TO
print('\nEqual to --> \n', h == i)

# NOT EQUAL TO
print('\nNot Equal To --> \n', h != i)

# GREATER THAN EQUAL TO
print('\nGreater than equal to --> \n', h >= i)

# LESS THAN EQUAL TO
print('\nLess than equal to --> \n', h <= i)
```

```python
import torch
import time

# Define the size of metrices
size = 20000                      # Large size for performance comparison

# Create Random matrices un CPU
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# Measure time to Complete task on CPU
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)   # Matrix Multiplication on CPU
cpu_time = time.time() - start_time

print(f'Time On CPU ==> {cpu_time: 4f} Seconds')


# Move Matrices to GPU
matrix_gpu1 = matrix_cpu1.to('cuda')
matrix_gpu2 = matrix_cpu2.to('cuda')

# Measure time to complete task on GPU
start_time1 = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)    # Matrix Multiplication on GPU
torch.cuda.synchronize()            # Ensure all GPU operations are complete
gpu_time = time.time() - start_time1

print(f'\nTime On GPU ==> {gpu_time: 4f} Seconds')

# Compare Results
print('\nSpeedup (CPU time / GPU time) : ', cpu_time / gpu_time)
```
